## Import Library

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Memuat gambar

```python
path_image = 'buah.jpg'
img = cv2.imread(path_image)
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Konversi ke HSV

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

## Rentang warna buah

```python
lower_frt = np.array([0, 50, 50])
upper_frt = np.array([30, 255, 255])
mask_frt = cv2.inRange(hsv, lower_frt, upper_frt)
```

## Rentang warna daun

```python
lower_leaf = np.array([30, 40, 40])
upper_leaf = np.array([90, 255, 255])
mask_leaf = cv2.inRange(hsv, lower_leaf, upper_leaf)
```

## Mask untuk buah dan daun

```python
frt_just = cv2.bitwise_and(img, img, mask=mask_frt)
leaf_just = cv2.bitwise_and(img, img, mask=mask_leaf)
```

## Menampilkan hasil

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].imshow(image_rgb)
axs[0, 0].set_title('Gambar Asli')
axs[0, 0].axis('on')

axs[0, 1].imshow(mask_frt, cmap='gray')
axs[0, 1].set_title('Mask Buah')
axs[0, 1].axis('on')

axs[1, 0].imshow(cv2.cvtColor(frt_just, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('Segmentasi Buah')
axs[1, 0].axis('on')

axs[1, 1].imshow(cv2.cvtColor(leaf_just, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title('Segmentasi Daun')
axs[1, 1].axis('on')

plt.tight_layout()
plt.show()
```
